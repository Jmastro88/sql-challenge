```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import psycopg2 
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:4903Uci*@localhost:5432/employee_db')
connection = engine.connect()
```

```python
employees=  pd.read_sql('Select * from employees', connection)
```

```python
employees.head()
```

```python
salaries= pd.read_sql('Select * from salaries', connection)
```

```python
salaries
```

```python
max_sal=salaries['salary'].max()
min_sal=salaries['salary'].min()

print(f"Max Salary {max_sal}  Min Salary {min_sal}")
```

```python
plt.hist(salaries.salary) 

plt.title('Employee Salary Distribution')
plt.ylabel("Number of Employees With Salary")
plt.xlabel('Salary Ranges')
plt.grid()
```

```python
title=pd.read_sql('Select * from titles', connection)
title
```

```python
title_salary_combo=title.merge(salaries, on='emp_no', how='inner')

title_salary_combo
```

```python
grouped_sal=title_salary_combo.groupby("title")["salary"].mean()

grouped_sal.plot.bar('title','salary', color="pink")

plt.ylabel("Salary")

plt.show()
```
